```python
# base 4.00 , 0.25 per 140 m
def func(dist):
    base = 4.00
    return base + (dist//140)*0.25

for i in range(1, 10000, 100):
    print(func(i))
```

## 84

```python
import random
def func(a,b,c):
    med = a if a>b else b
    med = med if med < c else c
    if med == c and a > b:
        med = c if c > b else b
        return med
    elif med == c and a<b:
        med = c if c > a else a
        return med
    return med
    
l = list(range(1,1002))
random.shuffle(l)
for i in range(len(l) -3) :
    print(l[i],l[i+1], l[i+2], ' ' ,func(l[i],l[i+1],l[i+2]))
```

```python
num = list(range(1,13))
random.shuffle(num)
def fun(n):
    l = ['first','second','third','fourth','fifth','sixth','seventh','eighth','nineth','tenth','eleven','twelve']
    return l[n-1]
for i in num : 
    print(i, ' : ' ,fun(i))
```

## 88

```python
l = list(range(1,1000))
random.shuffle(l)
def func(a,b,c):
    k = [a,b,c]
    k.sort()
    t = True if k[0] + k[1] > k[2] else False
    return t
for i in range(len(l)-3) :
    print(l[i],l[i+1],l[i+2])
    print(func(l[i],l[i+1],l[i+2]))

    
```

```python
while True :
    b = input()
    if b == '':
        break
    a = int(b)
    for i in range(2, int(a/2) +1):
        if a  % i is 0 :
            print(a, 'is not prime number')
            break
    else :
        print(a,'is prime number' )
```
